```python
class State:
    def __init__(self, Amogh, Ameya, GrandMother, GrandFather, Time_taken, umbrella):
        self.Amogh = Amogh
        self.Ameya = Ameya
        self.GrandMother = GrandMother 
        self.GrandFather = GrandFather
        self.Time_taken = Time_taken
        self.umbrella = umbrella
    def goalTest(self):
        if self.Amogh == self.Ameya == self.GrandMother  == self.GrandFather == 'R':
            return True 
        else:
            return False
    def moveGen(self):
        children = []
        people = ['Amogh', 'Ameya', 'GrandMother', 'GrandFather']
        times = {'Amogh': 5, 'Ameya': 10, 'GrandMother': 20, 'GrandFather': 25}
        current_side = []
        for p in people:
            if p=="Amogh" and self.Amogh == self.umbrella:
                current_side.append(p)
            if p=="Ameya" and self.Ameya == self.umbrella:
                current_side.append(p)
            if p=="GrandMother" and self.GrandMother == self.umbrella:
                current_side.append(p)
            if p=="GrandFather" and self.GrandFather == self.umbrella:
                current_side.append(p)
        if self.umbrella == 'L':
            for i in range(len(current_side)):
                for j in range(i + 1, len(current_side)):
                    p1, p2 = current_side[i], current_side[j]
                    new_pos={}
                    for k in people:
                        if k in [p1, p2]:
                            new_pos[k] = 'R'
                        else:
                            if k=="Amogh":
                                new_pos[k] = self.Amogh
                            elif k=="Ameya":
                                new_pos[k] = self.Ameya
                            elif k=="GrandMother":
                                new_pos[k] = self.GrandMother
                            elif k=="GrandFather":
                                new_pos[k] = self.GrandFather
                    new_time = self.Time_taken + max(times[p1], times[p2])
                    new_state = State(new_pos["Amogh"],new_pos["Ameya"],new_pos["GrandMother"],new_pos["GrandFather"], Time_taken=new_time, umbrella='R')
                    if new_state.isSafe():
                        children.append(new_state)
        else:
            for p in current_side:
                new_pos={}
                for k in people:
                        if k==p:
                            new_pos[k] = 'L'
                        else:
                            if k=="Amogh":
                                new_pos[k] = self.Amogh
                            elif k=="Ameya":
                                new_pos[k] = self.Ameya
                            elif k=="GrandMother":
                                new_pos[k] = self.GrandMother
                            elif k=="GrandFather":
                                new_pos[k] = self.GrandFather
                new_time = self.Time_taken + times[p]
                new_state = State(new_pos["Amogh"],new_pos["Ameya"],new_pos["GrandMother"],new_pos["GrandFather"], Time_taken=new_time, umbrella='L')
                new_pos = {k: ('L' if k == p else getattr(self, k)) for k in people}
                if new_state.isSafe():
                    children.append(new_state)
        return children
    def isSafe(self):
        if self.Time_taken>60:
            return False
        else:
            return True
    def __str__(self):
        return "Amogh: " + self.Amogh + " Ameya: " + self.Ameya + " GrandMother: " + self.GrandMother + " GrandFather: " + self.GrandFather + " Umbrella: " + self.umbrella + " Time: " + str(self.Time_taken)
    def __eq__(self, other):
        if self.Amogh == other.Amogh and self.Ameya == other.Ameya and self.GrandMother == other.GrandMother and self.GrandFather == other.GrandFather and self.umbrella == other.umbrella:
            return True
        else:
            return False
    def __hash__(self):
        return hash(self.Amogh+self.Ameya+self.GrandMother+self.GrandFather+self.umbrella)
def reconstructPath(goal_node_pair, CLOSED):
    #goal->parent->parent-> ..... ->parent->None
    parent_map = {}
    for node, parent in CLOSED:
        parent_map[node] = parent
    path = []
    goal_node, parent = goal_node_pair
    path.append(goal_node)
    while parent is not None:
        path.append(parent)
        parent = parent_map[parent]
    return path
def removeSeen(children, OPEN, CLOSED):
    open_nodes = [node for node, parent in OPEN]
    closed_nodes = [node for node, parent in CLOSED]
    new_nodes = [c for c in children if c not in open_nodes and c not in closed_nodes]
    return new_nodes
def bfs(start):
    OPEN = [(start, None)]
    CLOSED = []
    while OPEN:
        node_pair = OPEN.pop(0)
        N, parent = node_pair
        if N.goalTest():
            print("Goal is found")
            path = reconstructPath(node_pair, CLOSED)
            path.reverse()
            for node in path:
                print(node)
            return
        else:
            CLOSED.append(node_pair)
            children = N.moveGen()
            new_nodes = removeSeen(children, OPEN, CLOSED)
            new_pairs = [(node, N) for node in new_nodes]
            OPEN = OPEN + new_pairs
    return []
#DFS
def dfs(start):
    OPEN = [(start, None)]
    CLOSED = []
    while OPEN:
        node_pair = OPEN.pop(0)
        N, parent = node_pair
        if N.goalTest():
            print("Goal is found")
            path = reconstructPath(node_pair, CLOSED)
            path.reverse()
            for node in path:
                print(node)
            return
        else:
            CLOSED.append(node_pair)
            children = N.moveGen()
            new_nodes = removeSeen(children, OPEN, CLOSED)
            new_pairs = [(node, N) for node in new_nodes]
            OPEN = new_pairs + OPEN #LIFO
    return []
start_state = State(Amogh='L', Ameya='L', GrandMother='L', GrandFather='L', Time_taken=0, umbrella='L')
print("BFS")
bfs(start_state)
print("DFS")
dfs(start_state)
```
